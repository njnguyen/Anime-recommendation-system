## Popularity Based Recommendation System

```python
ratings_with_name = ratings.merge(animes, on='anime_id')
ratings_with_name
# rating_x is individual ratings given from users
# rating_y is the average rating from all users (ignored for this project)
```

```python
# Groups the anime names by the number of individual ratings they have and creates a table of the number of ratings
num_ratings_df = ratings_with_name.groupby('name').count()['rating_x'].reset_index()
num_ratings_df.rename(columns={'rating_x':'num_ratings'}, inplace=True)
num_ratings_df
```

```python
# Groups the anime names by the number of individual ratings they have and creates a table of the average ratings
avg_ratings_df = ratings_with_name.groupby('name').mean()['rating_x'].reset_index()
avg_ratings_df.rename(columns={'rating_x':'avg_ratings'}, inplace=True)
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on='name')
popular_df
```

```python
# Only considering animes with more than 1000 ratings
popular_df = popular_df[popular_df['num_ratings'] >= 1000].sort_values('avg_ratings', ascending=False)
```

```python
popular_df = popular_df.merge(animes, on='name')[['name', 'num_ratings', 'avg_ratings', 'genre', 'type', 'episodes', 'image_url']]
popular_df.head(50)
```

## Collaborative Filtering Based Recommendation System

```python
# Groups the same user id's and gives the users with over 200 ratings a True
# value, False otherwise
x = ratings_with_name.groupby('user_id').count()['rating_x'] > 200

# Filters out x to select only the elements with a True value and
# retrieves their indices
users = x[x].index
```

```python
# Filters ratings_with_name based on whether the user_id is in users
filtered_ratings = ratings_with_name[ratings_with_name['user_id'].isin(users)]
```

```python
# Groups the same show names with the filtered ratings
# and gives shows with >= 50 ratings a True value, False otherwise
y = filtered_ratings.groupby('name').count()['rating_x'] >= 50

# Filters out y to select only the elements with a True value and
# retrieves their indices
popular_shows = y[y].index
```

```python
# Filters filtered_ratings based on whether the name is in popular shows
final_ratings = filtered_ratings[filtered_ratings['name'].isin(popular_shows)]
```

```python
pt = final_ratings.pivot_table(index='name', columns='user_id', values='rating_x')
```

```python
# Creates a pivot table of show names and ratings from users
# Fills in no ratings from users with 0s
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Gets the cosine similarities for each entry in the pivot table
similarity_scores = cosine_similarity(pt)
```

```python
# Recommends the top 10 most similar animes to the inputed show
# Params:
#   anime_names: string -> Inputed show to find similarities
# Returns none
def recommend(anime_name):
    # Fetch index in the pivot table where the anime is
    index = np.where(pt.index == anime_name)[0][0]
    # Creates a list of indices and their similarity scores sorted from high to low
    # Skips first entry, bc that is the given show
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse=True)[1:11]
    
    """
    # Prints the names of the similar shows from their indices
    for i in similar_items:
        print(pt.index[i[0]])
    """
    data = []
    for i in similar_items:
        item = []
        temp_df = animes[animes['name'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('name')['name'].values))
        item.extend(list(temp_df.drop_duplicates('name')['genre'].values))
        item.extend(list(temp_df.drop_duplicates('name')['type'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Shokugeki no Souma')
```
